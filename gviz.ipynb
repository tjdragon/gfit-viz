```python
import folium
import pandas as pd 
from colour import Color

green = Color("green")
colors = list(green.range_to(Color("red"), 10))

# https://python-visualization.github.io/folium/modules.html#module-folium.vector_layers

trackpoints = pd.read_csv("oc.csv", sep=r'\s*,\s*', engine='python') 
maxSpeed = trackpoints["s"].max()
currentcol = green
currentRadius = 1

def col4speed(s):
    global currentRadius
    global currentcol
    if not pd.isna(s):
        idx = int(s * 9.0 / maxSpeed)
        currentRadius = idx + 1
        currentcol = colors[idx]

map = folium.Map(location=[1.3521, 103.8198])
folium.Marker(location=[1.3521, 103.8198], popup='SG').add_to(map)

for index, row in trackpoints.iterrows():
    _s = row['s']
    col4speed(_s)
    lat = row['lat']
    lon = row['lon']
    if not pd.isna(lat):
        folium.CircleMarker(location=(lat, lon),
                            radius = currentRadius,
                            color="%s" % currentcol,
                            popup="Start",
                            fill=False).add_to(map)   
        

map.save('gviz.html')
map
```
